```python
from gensim.models import Word2Vec
import pandas as pd
import numpy as np
import MeCab
```

# 分かち書き

```python
# 有価証券報告書のテキストデータを持ってくる
company_and_contents = pd.read_csv('first_selected_company.csv')
docs = company_and_contents['経営方針'].tolist()
```

```python
company_and_contents
```

```python
# mecabで分かち書きに

processed_docs = []

for doc in docs:
    m = MeCab.Tagger("-Ochasen")

    # 形態素解析の結果を格納するリスト
    nouns = []

    # 形態素解析の結果を解析
    node = m.parse(doc).split("\n")
    for i in node:
        if i == "EOS" or i == "":
            continue
        else:
            chunk = i.split("\t")
            pos = chunk[3].split("-")[0]  # 品詞情報がある部分を取得
            if pos == "名詞":  # 名詞のみを取得
                nouns.append(chunk[0])
    # 重要！！！！　freaquencyの時は下の１行を消す。tfidfの時は下の１行必要！！！doc2vecの時も必要！！テキストがカンマで区切られるか、それとも半角スペースで区切られるか
    # nouns = " ".join(nouns)

    processed_docs.append(nouns)

print(processed_docs[0])
```

```python
# サンプルデータ
# 各文は単語のリストとして与えられます
sentences = [
    ['I', 'love', 'machine', 'learning'],
    ['machine', 'learning', 'is', 'fun'],
    ['deep', 'learning', 'is', 'a', 'subfield', 'of', 'machine', 'learning'],
    ['word2vec', 'is', 'an', 'algorithm', 'used', 'in', 'machine', 'learning']
]

# Word2Vecモデルの設定と訓練
model = Word2Vec(processed_docs, vector_size=100, window=5, min_count=1, workers=4)

# モデルの保存
model.save("2nd_screening.model")

# 保存したモデルを読み込む
model = Word2Vec.load("2nd_screening.model")
```

```python
# 'machine' と最も類似する単語を探す
similar_words = model.wv.most_similar('インテグリティ', topn=100)
print(similar_words)

# 'machine' のベクトルを取得
vector = model.wv['誠実']
print(vector)
```

```python
print(similar_words[0])
print(type(similar_words[0][0]))
```

```python
keywords = []
for i in range(len(similar_words)):
    keywords.append(similar_words[i][0])
print(keywords)
```

```python
score = 0
score += np.dot(vector, model.wv['社風'])
print(score)
```

```python
company_and_contents['score'] = None
company_and_contents
```

```python
for m in range(len(processed_docs)):
    score = 0
    for n in range(len(keywords)):
        if keywords[n] in processed_docs[m]:
            score += np.dot(vector, model.wv[keywords[n]])
    company_and_contents.at[m, 'score'] = score
    
```

```python
company_and_contents
```

```python
sorted_company_and_contents = company_and_contents.sort_values(by='score', ascending=False)
sorted_company_and_contents
```

```python
sorted_company_and_contents.iloc[:100][['name', 'score']].reset_index(drop=True).to_csv('second_selected_company.csv')
```
