```python
import gensim
import csv
import pandas as pd
```

```python
model = gensim.models.KeyedVectors.load('chive-1.2-mc30.kv')
```

```python
match = model.most_similar("誠実", topn=300)
```

```python
print(match)
keywords = []
for i in match:
    keywords.append(i[0])
```

```python
print(keywords)

keywords = pd.DataFrame(keywords)
```

```python
keywords.to_csv("keywords_word2vec.csv")
```
