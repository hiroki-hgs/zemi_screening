```python
#インテグリティ関連単語を決めるためのコード。指定文書の単語を出現数順に並べて表示した。

import MeCab
import os
import csv
tagger = MeCab.Tagger()
```

```python
# テキストファイルを読み込む
with open('what_is_integrity.txt', 'r') as file:
    text = file.read()
```

```python
noun_count = {}

node = tagger.parseToNode(text)
while node:
    word = node.surface
    hinshi = node.feature.split(",")[0]
    if word in noun_count.keys() and hinshi == "名詞":
        noun_freq = noun_count[word]
        noun_count[word] = noun_freq + 1
    elif hinshi == "名詞":
        noun_count[word] = 1
    else:
        pass
    node = node.next
    
noun_count = sorted(noun_count.items(), key=lambda x:x[1], reverse=True)
```

```python
print(noun_count)
```

```python
keywords = [item[0] for item in noun_count]

print(keywords)
```

```python
print(type(keywords))
```

```python
# 出力先のディレクトリとファイル名を指定
output_dir = os.path.join("..", "datasets")
output_file = os.path.join(output_dir, "keywords_mecab.csv")

# ディレクトリが存在しない場合は作成
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# CSVとして出力
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for item in keywords:
        writer.writerow([item])
```
