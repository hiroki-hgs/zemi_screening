```python
import pandas as pd
```

東証プライム上場企業の株式コードと株式の収益率を取得する

```python
df202204 = pd.read_csv('202204.csv', encoding='shift_jis')
df202204 = df202204[df202204.iloc[:, 4].notna()]

df202303 = pd.read_csv('202303.csv', encoding='shift_jis')
df202303 = df202303[df202303.iloc[:, 4].notna()]
```

```python
df202204
```

```python
df202303
```

```python
# まず、df202303の2列目の値のリストを取得します。
values_to_keep = df202303.iloc[:, 1].tolist()

# 次に、df202204の2列目の値が上記のリストに存在しない行を削除します。
df202204 = df202204[df202204.iloc[:, 1].isin(values_to_keep)]

```

```python
df202303
```

## 

```python
# まず、df202204の2列目の値のリストを取得します。
values_to_keep = df202204.iloc[:, 1].tolist()

# 次に、df202303の2列目の値が上記のリストに存在しない行を削除します。
df202303 = df202303[df202303.iloc[:, 1].isin(values_to_keep)]
```

```python
df202204
```

```python
df202303
```

```python
# df202204とdf202303を2列目の値（code）に基づいてマージ
df_merged = pd.merge(df202204.iloc[:, [1, 4]], df202303.iloc[:, [1, 4]], on=df202204.columns[1], how='inner')

# 新しい列名を設定
df_merged.columns = ['code', '202204', '202303']
```

```python
df_merged
```

```python
# 計算を行う
df_merged['return'] = (df_merged['202303'] - df_merged['202204']) / df_merged['202204']
# 指定した列を削除
df_merged = df_merged.drop(['202204', '202303'], axis=1)
```

```python
df_merged
```

```python
df_merged['code'] = df_merged['code'] * 10
df_merged
```
